## In this We are Extracting the features names

	1.In the from the Data columns and in the if condition we are conditioning that the features of data set columns
    2.if it is null and do the sum of it and the sum is greater then one . 

```python
features_with_nan=[features for features in df.columns if df[features].isnull().sum()>=1]
```

```python
features_with_nan
```

```python
for features in features_with_nan:
    print(features,np.round(df[features].isnull().mean()*100,4), '% missing values')
    #print(feature,np.round(dataset[feature].isnull().mean()*100,5), '% missing values')
```

```python
df1=df.copy()
for features in features_with_nan:
    df1[features]=np.where(df1[features].isnull(),1,0)
   	# Lets calculate the mean SalePrice where the information is missing or present
    df1.groupby(features)['SalePrice'].median().plot.bar()
    plt.title(features)
    plt.show()
```

#### Relationship between discrete variables and saleprice

```python
df2=df.copy()
for feature in Dis_feature:
    df2.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
    
```

#### Relationship between continous and saleprice

```python
df2=df.copy()
for feature in cont_feature:
    df2.groupby(feature)['SalePrice'].median().plot.hist()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
df2=df.copy()
for feature in cont_feature:
    df2[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```
